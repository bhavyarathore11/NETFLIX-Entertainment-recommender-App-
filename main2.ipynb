```python
tv_shows=tv_shows[['id', 'title', 'overview', 'genres']]
```

```python
tv_shows
```

```python
tv_shows['tags'] = tv_shows['overview']+tv_shows['genres']
```

```python
tv_shows
```

```python
tv_shows = tv_shows.dropna()
```

```python
tv_shows
```

```python
new_data  = tv_shows.drop(columns=['overview', 'genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="Stranger Things"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(tv_shows):
    index=new_data[new_data['title']==tv_shows].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand("Stranger Things")
```
